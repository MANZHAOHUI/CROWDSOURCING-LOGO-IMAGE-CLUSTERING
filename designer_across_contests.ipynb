```python
from sklearn.metrics import silhouette_score
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

def get_data(path):
    
    filenames = os.listdir(path)
    path_filenames = []
    filename_list = []
    for file in filenames:
        if not file.startswith('.'):
            path_filenames.append(os.path.join(path, file))
            filename_list.append(file)
    features = []
    files = path_filenames

    sift = cv2.xfeatures2d.SIFT_create()
    for file in files:
        
        try:
            
            img = cv2.imread(file)

            nhist = cv2.calcHist([img], [0,1], None, [64,64], [0.0, 256.0,0.0, 256.0])

            nhist = cv2.normalize(nhist, nhist, 0, 256, cv2.NORM_MINMAX).flatten()
        except:
            print (file)
        
        features.append(nhist)
#         img=cv2.imread(file, 0) 
#         img=cv2.resize(img,(64,64),interpolation=cv2.INTER_CUBIC)

#         #创建二维列表
#         h, w = img.shape[:2]
#         vis0 = np.zeros((h,w), np.float32)
#         vis0[:h,:w] = img       #填充数据

#     #二维Dct变换
#         vis1 = cv2.dct(cv2.dct(vis0))
#     #cv.SaveImage('a.jpg',cv.fromarray(vis0)) #保存图片
#         vis1.resize(32,32)

#     #把二维list变成一维list
#         img_list=flatten(vis1.tolist())
        
#         features.append(img_list)

    input_x = np.array(features)
    data=input_x

    return data

def optimalK(data,maxClusters):
    
    cluster_num=[]
    sil_score=[]
        # Fit cluster to original data and create dispersion
    if maxClusters>2:
        
        for n_cluster in range(2, maxClusters):
            try:
                kmeans = KMeans(n_clusters=n_cluster).fit(data)
                label = kmeans.labels_
                sil_coeff = silhouette_score(data, label, metric='euclidean')
            except:
                print (n_cluster)
            cluster_num.append(n_cluster)
            sil_score.append(sil_coeff)
            #print("For n_clusters={}, The Silhouette Coefficient is {}".format(n_cluster, sil_coeff))
        return sil_score.index(max(sil_score))+2
    else:
        return 1


```

```python
def get_diversity (folder_name, img_num):
    path_folder='c:\\cropped\\img_d/'+folder_name+'/'
    data = get_data(path_folder)
    m=optimalK(data, img_num)
    return m


dv=pd.DataFrame(pd.read_csv('c:\\cropped/designer_en_num.csv'))
```
